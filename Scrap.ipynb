```python
import json
def Cleaning_the_files(jobs):
    data = jobs
    data = [job for job in data if ",\n," not in job]
    unique_data = [dict(t) for t in {tuple(d.items()) for d in data}]
    file_path = 'Jobs.txt'     # to save as text files
    with open(file_path, 'w', encoding='utf-8') as file:
        for job in unique_data:
            file.write(json.dumps(job) + ',')

    file_path = 'jobs.json'      # to save Json files
    def download_json_file(unique_data, file_path):
        with open(file_path, 'w', encoding='utf-8') as file:
            json.dump(unique_data, file, indent=4)
    download_json_file(unique_data, file_path)
Cleaning_the_files(jobs)
```
