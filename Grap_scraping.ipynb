```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://adrevenueindex.ezoic.com/"
response = requests.get(url)
html_content = response.content
```

```python
soup = BeautifulSoup(html_content, "html.parser")
```

```python
graph_div = soup.find("div", class_="highcharts-container")
if graph_div is not None:
    index_numbers = graph_div.find_all("text", class_="highcharts-axis-labels highcharts-yaxis-labels")
    for index_number in index_numbers:
        value = index_number.get_text()
        print(f"Index Number: {value}")
else:
    print("Graph element not found.")
```

```python
import csv
with open("data.csv", "w", newline="") as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(["Date", "Index Number"])  # Write header row
    for index_number in index_numbers:
        date = index_number.get("data-date")
        value = index_number.get_text()
        writer.writerow([date, value])
```

```python
import requests

url = "https://niche-trends-index-backend.ezoic.com/adindex/series/United%20States"

try:
    response = requests.get(url)
    response.raise_for_status()
    content = response.content.decode("utf-8")

    # Find the index where "categories: 1633305600000" starts
    start_index = content.find("categories: 1633305600000")

    if start_index != -1:
        # Extract the data starting from the found index
        data = content[start_index:]
        print(data)
    else:
        print("Data not found")
except requests.exceptions.RequestException as e:
    print("Request failed:", e)
```
